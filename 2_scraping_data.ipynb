# Scraping the Wikipedia page

```python
import pandas as pd
```

```python

url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

df=pd.read_html(url, header=0)[0]

df.head()
```

## Ignore cells with a borough that is Not assigned.

```python
delete_row = df[df['Borough'] == 'Not assigned'].index
df = df.drop(delete_row)
df.head()
```

##  More than one neighborhood can exist in one postal code area.

```python
#Count total values
df['Postal code'].size
```

```python
#Count unique values
df['Postal code'].nunique()
```

There are no more than one neighborhood in one postal code area!

## Neighborhoods separated with a comma 

There are same Neighbourhoods which are separated with "/" like "Regent Park / Harbourfront". I am going to separate them with comma as required.

```python
df['Neighborhood']=df['Neighborhood'].str.replace(' /', ',')
```

```python

df.head()
```

## If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df.isna().sum()
```

There are no cells that has a borough but a Not assigned neighborhood!
